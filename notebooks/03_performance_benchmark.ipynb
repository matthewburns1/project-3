```python
import numpy as np
import time
import matplotlib.pyplot as plt

# Matrix size - same as C++
n = 200

# Create two random matrices
A = np.random.randint(0, 100, (n, n)).astype(float)
B = np.random.randint(0, 100, (n, n)).astype(float)

# Time Python with loops (fair comparison to C++)
start = time.time()
C = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        for k in range(n):
            C[i][j] += A[i][k] * B[k][j]
end = time.time()

python_time = end - start 
print(f"Matrix size: {n}x{n}")
print(f"Python execute time: {python_time:.4f} seconds")
```

```python
# Time NumPy - optimized Python
start = time.time()
C_numpy = np.dot(A, B) 
end = time.time()

numpy_time = end - start
print(f"NumPy execution time: {round(numpy_time, 6)} seconds")
print(f"\nSpeedup summary:")
print(f"C++ vs Python loops: {round(python_time/0.0970, 1)}x faster")
print(f"NumPy vs Python loops: {round(python_time/numpy_time, 1)}x faster")
print(f"C++ vs NumPy: {round(numpy_time/0.0970, 1)}x comparison")
```

```python
languages = ['Python\nLoops', 'C++', 'NumPy']
times = [python_time, 0.0970, numpy_time]
colors = ['red', 'steelblue', 'green']

plt.figure(figsize=(8, 5))
plt.bar(languages, times, color=colors, edgecolor='black')
plt.title('Matrix Multiplication Performance Comparison\n200x200 Matrix')
plt.ylabel('Execution Time (seconds)')
plt.ylabel('log')
plt.tight_layout()
plt.savefig('../outputs/performance_benchmark.png')
plt.show()

print("Chart saved!")
```
